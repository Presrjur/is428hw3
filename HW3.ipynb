```python
import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Load the data from CSV
df = pd.read_csv('BMI.csv')

# Define the features (excluding the target 'fatpctg')
features = df.columns[:-1]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df['fatpctg'], test_size=0.2, random_state=42)

# Function to perform forward stepwise regression
def forward_selection(X, y):
    # Initial model with no predictors
    predictors = []
    best_score = float('inf')

    # Iterate through features and add the one that improves the model the most
    for feature in X.columns:
        model = sm.OLS(y, sm.add_constant(X[predictors + [feature]])).fit()
        score = mean_squared_error(y, model.predict(sm.add_constant(X[predictors + [feature]])))

        if score < best_score:
            best_score = score
            predictors.append(feature)

    return predictors, best_score

# Perform forward stepwise regression
selected_features, best_score = forward_selection(X_train, y_train)

print(f"Selected features: {selected_features}")
print(f"Best MSE on training set: {best_score}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the data
# Assuming log_returns_df is the DataFrame containing log returns

# Derive log returns from the dataset
log_returns = pd.DataFrame()
for column in log_returns_df.columns[1:]:
    log_returns[column] = np.log(log_returns_df[column] / log_returns_df[column].shift(1))

# Drop the first row with NaN values
log_returns = log_returns.dropna()

# Standardize the data
scaler = StandardScaler()
X_standardized = scaler.fit_transform(log_returns)

# Fit a PCA model
pca = PCA()
principal_components = pca.fit_transform(X_standardized)

# Plot a scree plot
eigenvalues_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(eigenvalues_ratio)

plt.figure(figsize=(12, 6))
plt.bar(range(1, len(eigenvalues_ratio) + 1), eigenvalues_ratio, alpha=0.8, align='center', label='Individual variance')
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--', color='r', label='Cumulative variance')
plt.title('Scree Plot')
plt.xlabel('Principal Component Index')
plt.ylabel('Variance Ratio')
plt.legend()
plt.show()

# Create a second plot showing cumulative variance retained for top N components
plt.figure(figsize=(12, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--', color='r')
plt.title('Cumulative Variance Retained')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Variance (%)')
plt.show()

# Identify the number of components needed to capture 80% of the total variance
n_components_80_percent = np.argmax(cumulative_variance >= 0.8) + 1
print(f"Number of components to capture at least 80% of variance: {n_components_80_percent}")

# Plot the time series of the 1st principal component
pc1 = principal_components[:, 0]
plt.figure(figsize=(12, 6))
plt.plot(log_returns_df['date'], pc1, label='1st Principal Component')
plt.title('Time Series of 1st Principal Component')
plt.xlabel('Date')
plt.ylabel('Principal Component Value')
plt.legend()
plt.show()

# Identify the date with the lowest value for the 1st principal component
min_index = np.argmin(pc1)
date_lowest_value = log_returns_df['date'].iloc[min_index]
print(f"Date with the lowest value for the 1st principal component: {date_lowest_value}")
```
